```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```
tips = sns.load_dataset('tips')
```

```
tips.head()
```

```
tips.shape
```

```
sns.countplot(tips.day)
plt.show()
```

```
sns.countplot(tips.day, palette=["b", "r", "pink", 'skyblue'])
```

```
#how we can check
sns.countplot(tips.smoker)
```

```
sns(x='smoker, data')
```

```
#barplot with tip by day
sns.barplot(x='day', y='tip', data=tips)
```

```
sns.barplot(x='day', y='tip', ci=None, data=tips)
```

```
sns.set(rc={'figure.figsize': (12,8)})
sns.barplot(x='day', y='tip', ci=None, data=tips)
```

```
sns.reset_orig
sns.barplot(x='day', y='tip', ci=None, data=tips)
plt.show()
```

```
sns.set_style('darkgrid')
sns.set_palette('RdBu')
sns.set_context('poster')
sns.barplot(x='day', y='tip', ci=None, data=tips)
```

```
sns.reset_orig
sns.scatterplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```
sns.scatterplot(x='total_bill', y='tip', hue='smoker', hue_order = ['Yes', 'No'], data=tips)
```

```
plt.figure(figsize=(20, 10))
sns.scatterplot(x='total_bill', y='tip', hue='smoker', hue_order = ['Yes', 'No'], data=tips)
```

```
plt.figure(figsize=(20, 10))
sns.scatterplot(x='total_bill', y='tip', hue='smoker', hue_order = ['Yes', 'No'], 
                size='day', data=tips)
```

```
plt.figure(figsize=(20, 10))
sns.scatterplot(x='total_bill', y='tip', 
                hue='smoker', hue_order = ['Yes', 'No'], 
                size='day', style='sex'
                data=tips)
```

```
sns.relplot(x='total_bill', y='tip', data=tips)
```

```
sns.relplot(x='total_bill', y='tip', data=tips,
            kind='scatter',
            col='smoker',
            row='time')
```

```
path = 'https://raw.githubusercontent.com/RamiKrispin/coronavirus/master/csv/coronavirus.csv'
df=pd.read_csv(path)
df.shape
```

```
df.info
```

```
sns.countplot(x=df.type)
```

```
sns.barplot(x='country', y='cases', data=df)
```

```
sns.set(rc={'figure.figsize': (40, 20)})
sns.barplot(x='country', y='cases', ci=None, data=df)
plt.xticks(rotation=90)
plt.show()
```

```
sns.set(rc={'figure.figsize': (40, 20)})
sns.barplot(x='date', y='cases', ci=None, data=df)
plt.xticks(rotation=90)
plt.show()
```

```
df1=df[ (df['country']=='US') | (df['country']=='Russia') | (df['country']=='Brazil')]  #'|' means 'or'; '&' means 'and'
df1.shape
```

```
df1.head()
```

```
sns.lineplot(x='date', y='cases', hue='country', ci=None, data=df1)
plt.xticks(rotation=90)
```

```
sns.reset_orig()
sns.barplot(x='country', y='cases', hue='type', ci=None, data=df1)
plt.xticks(rotation=90)
plt.show()
```

```
sns.relplot(x='date', y='cases', row='country',
            height=5,
            aspect=5,
            ci=None, kind='line', data=df1)
plt.xticks(rotation=90)
plt.show()
```
